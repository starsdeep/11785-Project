```python
with open('ava_action_list_v2.1.pbtxt') as fd:
    lines = fd.readlines()
len(lines)

labels = []
for i in range(80):
    name_line = lines[i * 5 + 1]
    label_id_line = lines[i * 5 + 2]
    label_type_line = lines[i * 5 + 3]    
    
    name = name_line[name_line.find('"')+1:name_line.rfind('"')]
    label_id = int(label_id_line.strip().split(':')[1].strip())
    label_type = label_type_line.strip().split(':')[1].strip()
    
    assert label_id == i+1
    labels.append({
        'name':name,
        'label_type':label_type
    })
```

```python
from pprint import pprint
pprint(labels)
```
